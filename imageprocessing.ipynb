```python
%pip install Pillow
%pip install rembg
!pip install onnxruntime
# vb seda vaja pip install rembg==2.0.54  --no-deps --prefer-binary
import onnxruntime
from rembg import remove 
from PIL import Image 
import os

def format_folder_images(input_folder, output_folder_name="processed_images"):

    try:
        output_folder = os.path.join(input_folder, output_folder_name)
        os.makedirs(output_folder, exist_ok=True)
        for filename in os.listdir(input_folder):
            file_path = os.path.join(input_folder, filename)
            if not os.path.isfile(file_path) or not filename.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif')):
                continue
            img = Image.open(file_path)
            img_resized = img.resize((100, 100), Image.Resampling.LANCZOS)
            output_path = os.path.join(output_folder, os.path.splitext(filename)[0] + ".png")
            img_resized.save(output_path, format="PNG")
            print(f"Processed and saved: {output_path}")
        print(f"All images processed and saved in: {output_folder}")

    except Exception as e:
        print(f"Error processing images: {e}")
```
